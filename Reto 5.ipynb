## Ejercicio 1
Una función matemática para calcular el volumen y el área superficial.
Cree dos funciones en python para calcular los valores antes establecidos, al ingresar por teclado r1, r2 y h.
Revise como utilizar el valor de pi usando import math y math.pi

```python
import math
def vol_área(r1, r2, h):
    #Vamos a calcular primero el área total
    #Se calcula el área del círculo y del cono (área base + lateral)
    acir =  math.pi * (r1**2)
    abcon = math.pi * (r2**2)
    g = math.sqrt((h**2)+(r2**2))
    alcon = math.pi * r2 * g
    acon = abcon + alcon
    #Suma de áreas
    at = acir + acon

    #Calculamos ahora el volumen
    volci = (4*math.pi*(r1**3))/3
    volco = abcon*h/3
    vt = volci + volco

    print ("El área total de la figura es", at, "\n" , "El volumen total de la figura es ", vt)
```

## Ejercicio 2
Una función matemática para calcular el área y el perimetro.
Cree dos funciones en python para calcular los valores antes establecidos, al ingresar por teclado r, a y b.
Revise como utilizar el valor de pi usando import math y math.pi

```python
def áre_per(r, a, b):
    #Calculamos el perímetro del rectángulo y los 2 círculos y los sumamos
    pr = 2 * (a+b)
    pc = 4*math.pi*r
    pt = pr + pc

    #Calculamos el área total de las figuras
    ar = a*b
    ac= 2*math.pi*(r**2)
    at = ar+ac

    print("El perímetro total de la figura es", pt, "\nEl área total de la figura es", at)
```

## Ejercicio 4
Haga un programa que utilice una función para calcular el valor de un préstamo C usando interés compuesto del i por n meses.

```python
def préstamo(c, i, n):
    """
    Calcula el valor final de un préstamo usando interés compuesto.

    Args:
    c (float): Capital inicial (préstamo).
    i (float): Tasa de interés mensual (en forma decimal).
    n (int): Número de meses.

    Returns:
    float: Monto total a pagar.
    """
    # Fórmula del interés compuesto
    A = c * (1 + i)**n
    return A

# Solicitar datos al usuario
print("Cálculo del valor de un préstamo con interés compuesto:")
c = float(input("Ingrese el capital inicial (monto prestado): "))
i = float(input("Ingrese la tasa de interés mensual (en %): ")) / 100
n = int(input("Ingrese el número de meses: "))

# Calcular el monto total
monto_total = préstamo(c, i, n)

# Mostrar el resultado
print(f"\nEl monto total a pagar después de {n} meses es: {round(monto_total, 2)}")
```

## Reto 5
Escriba un programa que pida 5 números reales y calcule las siguientes operaciones usando una función para cada una:
- El promedio
- El promedio multiplicativo (multilplica todos y luego calcula la raíz de la cantidad de operandos)
- La potencia del mayor número elevado al menor número
- La raíz cúbica del menor número


```python
import math

numeros = []
#-------
def promedio(numeros):
  return sum(numeros) / len(numeros)

print("Ingresa 5 número reales: ")

for i in range(5):
  num = float(input("Ingrese el número:"))
  numeros.append(num)
promedio = promedio(numeros)

def promedio_multiplicativo(numeros):
  prod = math.prod(numeros)
  return prod **(1/len(numeros))

promedio_multiplo = promedio_multiplicativo(numeros)

def mayor_menor(numeros):
  mayor = max(numeros)
  menor = min(numeros)
  return mayor , menor
mayor, menor = mayor_menor(numeros)
potencia = mayor ** menor

def raiz_cubica(numeros):
  raiz = menor**(1/3)
  return raiz
raiz_menor = raiz_cubica(numeros)

print(f"El promedio es: {promedio}")
print(f"El promedio multiplicativo es: {promedio_multiplo}")
print(f"El mayor número elevado al menor es: {potencia}")
print(f"La raiz cubica del menor es: {raiz_menor}")
```

Pip es el administrador de paquetes para Python. Su nombre proviene de "Pip Installs Packages" y es una herramienta que permite instalar, actualizar y gestionar bibliotecas y paquetes de software escritos en Python.

- ¿Qué es pip?
Es un programa que se usa desde la línea de comandos (terminal o consola).
Permite descargar e instalar paquetes desde el Python Package Index (PyPI), un repositorio oficial que contiene miles de paquetes creados por la comunidad de Python.
También puede instalar paquetes desde otras fuentes, como repositorios privados o directamente desde un archivo local.
